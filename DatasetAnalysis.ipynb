# Convert Order Date and Ship Date into datetime format.

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

# Create Order Year, Order Month, Order Day, Ship Year and Ship Month columns.

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month']=df['Order Date'].dt.month
df['Order Day']=df['Order Date'].dt.day
df['Ship Year']=df['Ship Date'].dt.year
df['Ship Month']=df['Ship Date'].dt.month
```

# Calculate shipping time (days) for every order.

```python
df['Shipping Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

# Find the average shipping time.

```python
df['Shipping Time'].mean()
```

# Matplotlib: Bar chart of Sales by Category.


```python
import matplotlib.pyplot as plt
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

# Matplotlib: Bar chart of Profit by Region

```python
Region_Based_profit=df.groupby('Region')['Profit'].sum()
plt.bar(Region_Based_profit.index,Region_Based_profit.values)
plt.title("Profit By Region")
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

# Matplotlib: Horizontal bar chart of Top 10 Products by Sales.

```python

Top_Product = df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).head(10)

plt.barh(Top_Product.index , Top_Product.values)
plt.xlabel('Sales')
plt.ylabel('Product')

plt.title('TOP 10 Products by sales')
```

# Box Plot Sales

```python
plt.figure(figsize=(6,6))

plt.boxplot(
    df['Sales'],
    patch_artist=True
)

plt.title('Box Plot of Sales')
plt.ylabel('Sales')

plt.grid(axis='y', linestyle='-', alpha=0.4)

plt.show()
```

# Line chart Of Monthly Sales

```python
Monthly_Sales=df.groupby('Order Month')['Sales'].sum()
plt.figure(figsize=(12,6))
plt.plot(Monthly_Sales.index,Monthly_Sales.values,marker='D',linestyle='--')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales')

plt.grid(True)

plt.show()
```

# Correlation Heatmap By Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Matrix')
plt.show()
```

# Scatter plot of Sales vs Profit (hue=Category) By using Seaborn

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Sales',y='Profit')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```
